```python
from keras.datasets import mnist
import matplotlib.pyplot as plt
import numpy as np

(x_train, y_train), (x_test, y_test) = mnist.load_data()

#shuffling the training set
num_train_img=x_train.shape[0]
train_ind=np.arange(0,num_train_img)
train_ind_s=np.random.permutation(train_ind)
x_train=x_train[train_ind_s,:,:]
y_train=y_train[train_ind_s]

#Validation set
x_valid=x_train[0:int(0.2*num_train_img),:,:]
y_valid=y_train[0:int(0.2*num_train_img)]

#Training set
x_train=x_train[int(0.2*num_train_img):,:,:]
y_train=y_train[int(0.2*num_train_img):]

#shuffling the testing set
num_test_img=x_test.shape[0]
test_ind=np.arange(0,num_test_img)
test_ind_s=np.random.permutation(test_ind)
x_test=x_test[test_ind_s,:,:]
y_test=y_test[test_ind_s]

print('The number of images in training set is', x_train.shape[0])
print('The number of images in validation set is',x_valid.shape[0])
print('The number of images in testing set is',x_test.shape[0])
print('The image dimensions are', x_train.shape[1], 'x', x_train.shape[2])
#10 random images for training
num=10
num_row = 2
num_col = 5
fig, axes = plt.subplots(num_row, num_col, figsize=(1.5*num_col,2*num_row))
digit=0
i = 0
print("Selecting 10 random images from training set")
for i in range(num):
  x_train_d=x_train[y_train==digit,:,:]
  x_train_i=x_train_d[10,:,:] 
  ax = axes[i//num_col, i%num_col]
  ax.imshow(x_train_i, cmap='gray')
  ax.set_title('Label: ' + str(i))
  digit = digit + 1
  i = i +1
plt.tight_layout()
plt.show()

#10 random images for testing
num = 10
num_row = 2
num_col = 5
fig, axes = plt.subplots(num_row, num_col, figsize =(1.5*num_col,2*num_row))
digit=0
i = 0
print("Selecting 10 random images from testing set")
for i in range(num):
  x_test_d=x_test[y_test==digit,:,:]
  x_test_i=x_test_d[10,:,:] 
  ax = axes[i//num_col, i%num_col]
  ax.imshow(x_test_i, cmap='gray')
  ax.set_title('Label: ' + str(i))
  digit = digit + 1
  i = i +1
plt.tight_layout()
plt.show()

#10 random images for validation
num=10
num_row = 2
num_col = 5
fig, axes = plt.subplots(num_row, num_col, figsize=(1.5*num_col,2*num_row))
digit=0
i = 0
print("Selecting 10 random images from validation set")
for i in range(num):
  x_valid_d=x_valid[y_valid==digit,:,:]
  x_valid_i=x_valid_d[10,:,:] 
  ax = axes[i//num_col, i%num_col]
  ax.imshow(x_valid_i, cmap='gray')
  ax.set_title('Label: ' + str(i))
  digit = digit + 1
  i = i +1
plt.tight_layout()
plt.show()
```
